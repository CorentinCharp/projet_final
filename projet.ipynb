```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
```

```python
# Houses sold from 1995 and 2024, data given by the governement of the UK
house_pricing = pd.read_csv('kaggle_london_house_price_data.csv')
house_pricing
```

```python
house_pricing['fullAddress'].value_counts()
```

```python
house_pricing.describe()
```

```python
house_pricing.drop_duplicates()
```

```python
house_pricing.columns
```

```python
#On modifie le type p
house_pricing['history_date'] = pd.to_datetime(house_pricing['history_date'])

avg_prices = house_pricing.groupby(house_pricing['history_date'].dt.year)['history_price'].mean()
print(avg_prices)
```

```python
avg_prices.plot(kind='line', xlabel='years', ylabel ='price', title='Evolution of price of real estate in London through the years')
```
